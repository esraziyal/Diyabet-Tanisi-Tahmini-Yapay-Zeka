```python
# Gereken kütüphaneleri içeri aktarıyoruz
import pandas as pd
import numpy as np

# Aykırı Değerlerin Tespiti İçin IQR (Interquartile Range) Yöntemi kullandım.
def detect_and_fill_outliers(df):
    # Her bir sütun için IQR (İnter Çeyrek Aralığı) hesaplamak amacıyla, sadece sayısal veri türlerine sahip sütunları seçtim.
    for column in df.select_dtypes(include=[np.number]).columns:
        # 1. çeyrek (Q1, %25) ve 3. çeyrek (Q3, %75) değerlerini hesaplama
        Q1 = df[column].quantile(0.25)  # Q1, 25. persentil değeri
        Q3 = df[column].quantile(0.75)  # Q3, 75. persentil değeri

        # Çeyrekler arası fark (IQR) hesaplama: IQR = Q3 - Q1
        IQR = Q3 - Q1

        # Aykırı değerlerin alt ve üst sınırlarını belirleme:
        # Alt sınır: Q1 - 1.5 * IQR
        # Üst sınır: Q3 + 1.5 * IQR
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Aykırı değerlerin tespiti:
        # Eğer bir değeri, alt sınırdan küçük ya da üst sınırdan büyükse, bu değer aykırı kabul edilir.
        outliers = (df[column] < lower_bound) | (df[column] > upper_bound)

        # Aykırı değerleri doldurmak için, sütunun medyanını (orta değerini) kullanıyorum.
        # Medyan, aykırı değerlerden daha az etkilenir ve daha sağlıklı bir yerine koyma yöntemidir.
        median_value = df[column].median()

        # Aykırı olan değerleri medyan değeri ile dolduruyoruz
        # NumPy'nin np.where fonksiyonu ile, eğer değer aykırı ise medyan ile dolduruyorum, değilse orijinal değeri koruyorum.
        df[column] = np.where(outliers, median_value, df[column])

    # Temizlenmiş DataFrame'i geri döndürme işlemi
    return df

# less_data veri seti üzerinde aykırı değerleri tespit etme ve medyan ile doldurma işlemi:
less_data_cleaned = detect_and_fill_outliers(less_data)

# Temizlenmiş veri setinin ilk birkaç satırını görüntüleme işlemi:
print(less_data_cleaned.head())

#Bu kod, veri setindeki sayısal sütunlar üzerinde aykırı değerleri tespit edip, bu aykırı değerleri medyan ile doldurarak veri temizliği yapmayı amaçlamaktadır.
#Aykırı değerler, veri analizini ve modelleme süreçlerini olumsuz etkileyebilir. Bu nedenle, aykırı değerleri temizlemek modelin doğruluğunu artırabilir.

```

```python
# Gereken kütüphaneleri içeri aktarma
import matplotlib.pyplot as plt  # Grafik çizmek için matplotlib
import seaborn as sns  # Gelişmiş görselleştirmeler için
import numpy as np  # Sayısal işlemler için numpy kütüphanesi..

# Aykırı Değerler İçin Boxplot Çizimi (aykırı değerleri temizlediğimin kanıtı)
def boxplot_ciz(veri):
    # Grafik boyutlarını ayarlama
    plt.figure(figsize=(15, 10))  # Grafik boyutları (15x10 inç)

    # Sayısal sütunlar için boxplot çizimi
    # veri.select_dtypes(include=[np.number]) sadece sayısal (numeric) sütunları seçer.
    sns.boxplot(data=veri.select_dtypes(include=[np.number]), palette="coolwarm")

    # Grafik başlığı ve ayarları
    plt.title("Boxplot")  # Grafik başlığı
    plt.xticks(rotation=90)  # X eksenindeki sütun adlarını döndürme, böylece daha okunaklı oluyor
    plt.ylabel("Değerler")  # Y ekseni etiketini "Değerler" olarak belirleme
    plt.xlabel("Özellikler")  # X ekseni etiketini "Özellikler" olarak belirleme
    plt.tight_layout()  # Grafik elemanlarının çakışmasını engelleme, her şey düzgün yerleşsin diye.
    plt.show()  # Grafiğin ekranda gösterimi

# Boxplot'u çizdirme
boxplot_ciz(less_data)  # less_data veri setini fonksiyona gönderdik.


# Bu kod, verilen veri setindeki sayısal sütunlar için aykırı değerleri görselleştiren bir boxplot oluşturur. Aykırı değerler genellikle kutunun dışında bulunan noktalar olarak gösterilir.
# Bu görselleştirme, aykırı değerlerin veriden temizlenip temizlenmediğini ve verinin genel dağılımını görsel olarak incelememizi sağlıyor. Bu tür görselleştirmeler, verilerin doğruluğunu ve güvenilirliğini artırmak için önemlidir.


```

```python
# Gerekli kütüphaneleri ekleme
import pandas as pd
from sklearn.decomposition import PCA  # PCA'yı uygulamak için gerekli kütüphane
from sklearn.preprocessing import StandardScaler  # Veriyi ölçeklendirmek için gerekli kütüphane
import matplotlib.pyplot as plt  # Görselleştirme için gerekli kütüphane

# PCA için kullanılacak sayısal sütunları seçiyoruz
# Bu sütunlar, analiz için kullanılacak veriyi temsil eder
numerical_columns = ['HbA1c', 'Hba1c_change', 'age', 'LDL', 'Cholesterol',
                     'HDL', 'Creatinine', 'Triglyceride']

# Veriyi seçip sadece sayısal sütunları alıyoruz
# Burada sadece `numerical_columns` listesinde yer alan sütunlardan verileri alıyoruz
X = less_data[numerical_columns]

# Veriyi ölçeklendirme (standartlaştırma) işlemi
# PCA'nın düzgün çalışabilmesi için verilerin standartlaştırılması gerekiyor
# StandardScaler, her bir özelliği (sütunu) ortalama 0 ve standart sapma 1 olacak şekilde dönüştürür
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Veriyi ölçeklendiriyoruz

# PCA uygulama (Principal Component Analysis)
# Burada veriyi 2 bileşene indiriyoruz. Bu bileşenler, verinin en önemli yapısını yansıtır.
pca = PCA(n_components=2)  # 2 bileşene indirgeme
X_pca = pca.fit_transform(X_scaled)  # PCA'yı uyguluyoruz ve sonucu X_pca'ya atıyoruz

# PCA sonuçlarını görselleştirme
# Burada scatter plot ile PCA'nın ilk iki bileşenini görselleştiriyoruz
# `X_pca[:, 0]` 1. ana bileşeni, `X_pca[:, 1]` ise 2. ana bileşeni temsil eder
# `c=less_data['Glycemic_control']` ile noktaların rengini `Glycemic_control` değerine göre ayarlıyoruz
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=less_data['Glycemic_control'], cmap='jet', alpha=0.7)

# Grafik başlığını ve eksen etiketlerini ayarlıyoruz
plt.xlabel('1. Principal Component')  # 1. ana bileşenin adı
plt.ylabel('2. Principal Component')  # 2. ana bileşenin adı
plt.title('PCA - Glycemic Control Visualization')  # Grafik başlığı ekleme

# Renk çubuğunu ekliyoruz, burada renkler 'Glycemic_control' verilerine göre ayarlandı
plt.colorbar(label='Glycemic Control')

# Görseli görüntülüyoruz
plt.show()

# PCA açıklanan varyans oranını yazdırıyoruz
# Bu, her bir bileşenin verinin ne kadarını açıkladığını gösterir
print("Açıklanan varyans oranı:", pca.explained_variance_ratio_)

# Toplam açıklanan varyansı hesaplıyoruz
# Bu, tüm bileşenlerin veride açıklanan toplam varyansını gösterir
print("Toplam açıklanan varyans:", sum(pca.explained_variance_ratio_))


# Bu kod, Principal Component Analysis (PCA) kullanarak veri kümesindeki yüksek boyutlu verileri iki bileşene indirgemeyi ve ardından bu bileşenleri görselleştirmeyi amaçlamaktadır.
# PCA, veri kümesinin boyutunu düşürmek ve önemli özellikleri daha iyi görselleştirmek için kullanılır.
```
